```python
import pandas as pd
import os
import numpy
```

```python
def split_excel_by_track_id(input_file,path):
    # Load the Excel file into a pandas DataFrame
    df = pd.read_excel(input_file)
    
    # Get unique track_ids
    track_ids = df['Track ID'].unique()
    # Create a folder to store the split files
    output_track_id_folder = path
    output_track_id_file = os.path.join(output_track_id_folder, "track_ids.csv")
    numpy.savetxt(output_track_id_file, track_ids, delimiter=",")
    # Split the DataFrame and save each split into separate files
    for track_id in track_ids:
        
        # Create a folder to store the split files
        output_folder = path + "ids/"+ f'{track_id}'
        os.makedirs(output_folder, exist_ok=True)
        
        track_df = df[df['Track ID'] == track_id]
        
        output_file = os.path.join(output_folder, f"track_{track_id}.xlsx")
        track_df.to_excel(output_file, index=False)
        
        print(f"Split file saved: {output_file}")

# Example usage
path = './Data/pos_0/'
input_file = './Data/pos_0/pos_0_track_full.xlsx'  # Specify the input Excel file name
split_excel_by_track_id(input_file,path)
```
